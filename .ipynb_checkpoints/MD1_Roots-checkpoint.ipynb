# MƒÅjas darbs 1 - NelineƒÅro vienƒÅdojumu saknes

Lielgabala lode tiek iz≈°auta no zemes le≈Üƒ∑ƒ´ $ùúÉ$ ar sƒÅkuma ƒÅtrumu $v_0$. Lielgabala lodes augstumu $h(t)$ jebkurƒÅ brƒ´dƒ´ $t$ var modelƒìt ar ≈°ƒÅdu vienƒÅdojumu, pie≈Üemot, ka nav gaisa pretestƒ´bas:

$$ h(t) = v_0 sin(Œ∏)t - \frac{1}{2}gt^2 $$

Kur:
* $g$ ‚Äî brƒ´vas kri≈°anas paƒÅtrinƒÅjums, $g = 9,81 m/s^2$
* $Œ∏$ - ir le≈Üƒ∑is grƒÅdos
* $v_0$ - ir sƒÅkotnƒìjais ƒÅtrums, metros sekundƒì.

Varianti:

1. Bisection, $v_0=50m/s$, $Œ∏ = 45¬∞$, $H = 30m$
2. False-Position, $v_0=60m/s$, $Œ∏ = 30¬∞$, $H = 30m$
3. Secant method, $v_0=50m/s$, $Œ∏ = 55¬∞$, $H = 50m$

1.1. Definƒìt funkciju $h(t)$ un attƒìlot lodes augstuma izmai≈Üu laikƒÅ attiecƒ´bƒÅ pret zemi.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Nodefinƒìju konstantus mainƒ´gos, kas ir doti priek≈° variantiem
V_ZERO = 50
ANGLE_THETA = 45
GRAVITY = 9.81
TARGET_HEIGHT = 30

# Izveidoju funkciju h(t), kas veidota pƒìc aug≈°ƒÅ dotƒÅs formulas
def h(t):
    # ≈†eit es pƒÅrveidoju ANGLE_THETA uz radiƒÅniem, jo funkcija math.sin() sagaida input radiƒÅnos
    return V_ZERO * math.sin(math.radians(ANGLE_THETA)) * t - 0.5 * GRAVITY * t ** 2

# Izveidoju numpy array ar vƒìrtƒ´bƒÅm 0 lƒ´dz 10 ar soli 0.1, lai plot neb≈´tu saraustƒ´ts, ≈°ƒ´s b≈´tu laika vƒìrtƒ´bas funckijai
x_points = np.arange(0.0, 10.0, 0.1)
y_points = []

# Eju cauri katrai laika vƒìrtƒ´bai un aprƒìƒ∑inu lodes augstumu dotajƒÅ laika sprƒ´dƒ´, ja tas ir lielƒÅks par 0, pievienoju to y_points masƒ´vam
# lai netiktu saglabƒÅtas vƒìrtƒ´bas, kas teorƒìtiski nozƒ´mƒìtu, ka lode iet zemƒÅk par zemi
for t in x_points:
    height = h(t)
    if height > 0:
        y_points.append(height)

# PƒÅrveidoju x_points, lai nodzƒìstu tos laika sprƒ´≈æus, kad h bija mazƒÅks par 0
x_points_valid = x_points[:len(y_points)]

# Izveidoju plot ar punktiem
plt.plot(x_points_valid, y_points)
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.show()
```

1.2. Noskaidrot kopƒìjo laiku $T$, ko lode pavada gaisƒÅ.

```python
# https://en.wikipedia.org/wiki/Projectile_motion atradu ≈°o formulu
total_airtime = (2 * V_ZERO * math.sin(math.radians(ANGLE_THETA))) / GRAVITY
print(total_airtime)
```

1.4. Atrast maksimƒÅlo sasniegto augstumu $H_{max}$

```python
# https://en.wikipedia.org/wiki/Projectile_motion arƒ´ ≈°eit atradu formulu
max_height = (V_ZERO**2 * math.sin(math.radians(ANGLE_THETA)**2) / (2 * GRAVITY))
print(max_height)
```
