Raķete tiek palaista gaisā uz $45s$, un tās ātrumu laikā apraksta ar teorētisku modeli $v(t)$

$$ v(t) = 700 *ln(t + 1) - 2 t^2 + 30 t$$

Lidojuma laikā, iebūvētie sensori mēra raķetes faktisko ātrumu ar diskrētiem laika soļiem $dt = 2s$, šie rezultāti atšķiras no teorētiskā modeļa vērtībām par nelielu Δ.

$$ v_m(t) = v(t) + Δ $$

Jūsu uzdevums ir noskaidrot raķetes veikto distanci, pamatojoties uz mērījumiem, un aprēķināt distances kļūdu starp teorētisko modeli un mērījumiem.

Varianti:

1. Trapecveida likums, Simpsona 1/3
2. Trapecveida likums, Simpsona 3/8

```python
import numpy as np
import matplotlib.pyplot as plt

# Theoretical model
def v(t):
    return 700 * np.log(t + 1) - 2 * t**2 + 30 * t

# Measured model (will produce random results each time invoked)
def v_sensor(t):
    return v(t) - (np.random.rand(len(t)) * 300 - 200)

t0 = 0
t1 = 45

# Sensor measuring interval
dt = 2

# For plotting theoretical
t = np.linspace(t0, t1, 100)
v_theoretical = v(t)

# Timestamp for sensor (each 2s)
t_sensor = np.arange(t0, t1 + dt, dt)
v_measured = v_sensor(t_sensor)

plt.plot(t, v_theoretical, label='Theoretical')
plt.plot(t_sensor, v_measured,'r--', label='Measured')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.grid(True)
plt.show()
```
